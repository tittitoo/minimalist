```python
import PyPDF2
from pathlib import Path
import os

# Wanted file name
filename = 'output.pdf'
folder = 'Datasheet'
downloads_folder = os.path.join(os.path.expanduser('~'), 'Downloads')
directory = Path(downloads_folder, folder)

# Create a list of all PDF files in the directory
pdf_files = [f for f in os.listdir(directory) if f.endswith('pdf')]

# Sort the PDF files alphabetically
pdf_files.sort()

# Creat a PdfFileMerger object
merger = PyPDF2.PdfMerger()

# Add all the PDF files to the merger
for pdf_file in pdf_files:
    file_path = os.path.join(directory, pdf_file)
    merger.append(file_path)

# Write the ouput to a new PDF fiel
output_path = os.path.join(directory, filename)
with open(output_path, 'wb') as output_file:
    merger.write(output_file)

# Close the merger
merger.close()
```
