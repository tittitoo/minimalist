```python
#%%
import os
from pathlib import Path

import xlwings as xw

import hide
```

```python
# Where is the file? Also put in extension. Output will be Technical + file.
file = 'J12513 MODEC - BM-C-33 FPSO - NL WHISTLE B0.xlsx'
```

```python
downloads_folder = os.path.join(os.path.expanduser('~'), 'Downloads')
file_path = Path(downloads_folder, file)
try:
    new_book = xw.Book(file_path, password=hide.legacy)
except:
    new_book = xw.Book(file_path, password=hide.new)

new_book.sheets['Cover'].range('D39').value = 'TECHNICAL PROPOSAL'
new_book.sheets['Cover'].range('C42:C47').value = new_book.sheets['Cover'].range('C42:C47').raw_value
new_book.sheets['Cover'].range('D6:D8').value = new_book.sheets['Cover'].range('D6:D8').raw_value
new_book.sheets['Summary'].range('D20:F100').value = ''
new_book.sheets['Summary'].range('C20:C100').value = new_book.sheets['Summary'].range('C20:C100').raw_value
new_book.sheets['Summary'].range('G:K').delete()
skip_sheets = ['Config', 'Cover', 'Summary', 'Technical_Notes', 'T&C']
for sheet in new_book.sheet_names:
    ws = new_book.sheets[sheet]
    ws.range('A1').value = ws.range('A1').raw_value
    if sheet not in skip_sheets:
        last_row = ws.range('B100000').end('up').row
        ws.range('B3:B' + str(last_row)).value = ws.range('B3:B' + str(last_row)).raw_value
        ws.range('AM:AN').delete()
        ws.range('I:AK').delete()
        ws.range('F:G').delete()
new_book.sheets['Config'].delete()
new_book.sheets['T&C'].delete()
new_book.sheets['Summary'].activate()
new_book.save( downloads_folder + '/' + 'Technical ' + file[:-4] + 'xlsx', password='') 
```
